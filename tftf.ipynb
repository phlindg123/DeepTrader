```python
import numpy as np
import matplotlib.pyplot as plt
import gym
from IPython import display
#matplotlib inline
```

```python
env = gym.make("Taxi-v2")
action_size = env.action_space.n
state_size = env.observation_space.n

print(action_size, state_size)
```

```python
Q = np.zeros((state_size, action_size))
C = np.zeros((state_size, action_size))
pi = [1] + [0]*(state_size-1)
gamma = 0.95
def random_policy(env):
    S = (state_size,)
    A = env.action_space.n
    return np.ones(S + (A,))/A
def gen_seq(policy, env):
    S = (state_size,)
    A = env.action_space.n
    indices = np.arange(A)
    seq = []
    state = env.reset()
    while True:
        action = np.random.choice(indices, p=policy[state])
        next_state, reward, done, info = env.step(action)
        seq.append((state, action, reward))
        state = next_state
        if done:
            break
    return seq
```

```python
episodes = 100
T = 100
b = random_policy(env)

for e in range(episodes):
    seq = gen_seq(b, env)
    G = 0
    W = 1
    for st, at, rt in reversed(seq):
        G = gamma*G + rt
        C[st, at] = C[st, at] + W
        Q[st, at] = Q[st, at] + (W/C[st,at])*(G - Q[st, at])
        pi[st] = np.argmax(Q[st,:])
        if at != pi[st]:
            break
        W = W/b[st,at]


        

        
        
        
    
```

```python
state = env.reset()
while True:
    action = pi[state]
    state, reward, done, info = env.step(action)
    env.render()
    if done:
        env.close()
        break
    

```

```python
import tensorflow as tf
import tensorflow_probability as tfp
```

```python
sess = tf.Session()
mu = np.random.rand(3)
mu = tf.convert_to_tensor(mu, name="mu")


cov = np.random.rand(9).reshape(3,3)
cov = tf.convert_to_tensor(cov @ cov.T, name="cov")
norm = tfp.distributions.MultivariateNormalFullCovariance(loc=mu, covariance_matrix=cov)
log_q = norm.log_prob(norm.sample())
grad = tf.gradients(ys = log_q, xs = [mu])
sess.run(tf.global_variables_initializer())
sess.run(grad)
```
